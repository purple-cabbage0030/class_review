4. 논리 연산과 비교 연산
> 자바의 true/false 소문자 키워드 <br>
> 파이썬 True/False 키워드

```python
# =: 할당 연산자 / 대입 연산자
name = "바보"
print(name)
```

```python
# ==: 동등 비교 연산자 - 값이 동일할 경우에 True
# 실행 순서는 1==1을 비교 후 print()
print(1 == 1)
```

```python
# !=: 부정 연산자 - 값이 다를 경우에 True
print(1 != 1)
```

```python
# 문자열 타입 비교 가능한지! 가능하다.
print("a" == "A")
```
