```python
data = pd.read_csv('./dataset/국토안전관리원_위험요소프로파일_20201231.csv', encoding='euc-kr')
data
```

```python
df = pd.DataFrame({'count':data.groupby('원인').size()}).reset_index()
df
```

```python
# 경사도 불량으로 인한 사고가 가장 많았다.
df.sort_values('count', ascending=False)
df.sort_values(by=['count'], ascending=False).head()
# 2개의 결과 값이 같음 by= 가 생략가능해보임, 둘이 같은 결과 값이 나옴
```

```python
# 시설종류에 따라 책임자 추가
data['책임자'] = data['시설종류'].apply(lambda x : '이홍주' if x == '건축' else '정주영') 
data
```

```python
for i in data['구분']:
    print(str(i)[0:4])
```

```python
year=[]
for i in data['구분']:
    year.append(str(i)[0:4])
year
```

```python
# 사건 년도 추가
data['Year'] = year
data
```

```python
# 년도당 사건수
df2 = pd.DataFrame({'count':data.groupby('Year').size()}).reset_index()
df2
```

```python
# 책임자 데이터프레임 생성
data2 = pd.DataFrame({
    "책임자" : ['이홍주', '정주영','하예진'],
    "직위" : ['CEO', '직원', '이사'],
    "사고 후 처리" : ['감봉', '인사이동', '자진퇴사']
})
data2
```

```python
# merge 사용
data3 = pd.merge(left = data,
                right = data2,
                how = 'left',
                left_on='책임자',
                right_on='책임자')
data3
```

```python
data4 = pd.merge(left = data,
                right = data2,
                how = 'outer',
                left_on='책임자',
                right_on='책임자')
data4
```
