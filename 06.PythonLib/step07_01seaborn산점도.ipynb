변수(피처)들 의미

1. total_bill = 계산한 총 금액

2. tip = 팁 금액

3. sex = 성별

4. smoker = 흡연 여부

5. day = 요일

6. time = 시간대(팁 계산 시간)

7. size = 같이 식사한 인원

```python
# 레스토랑 팁 데이터셋 활용
'''
미션: tip을 어떤 사람이 많이 주는지?

'''

tips = sns.load_dataset("tips")
tips
```

```python
tips.columns
```

```python
# total_bill과 tip의 관계도 확인 - 두 개 데이터의 상관관계 볼 때 좋은 건 산점도

sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용, feature를 색으로 구분
# 흡연 여부에 따른 색을 구분 - 범례 표현도 자동으로 설정됨
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker');
```

```python
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size');
```

```python
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker');
```

```python
# 기본 색을 파란색으로 적용
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker', palette='Blues');
```

```python
sns.set_style(style='whitegrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker', size='size');
```
